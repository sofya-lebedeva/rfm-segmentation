# Вопрос №4
> *Выяви самые популярные предметы (ТОП-3) по количеству регистраций на них.

```python
st_reg
```

```python
count_reg = st_reg.groupby('code_module') \
.agg({'date_registration': 'count'}) \
.rename(columns={'date_registration': 'count_reg'}) \
.reset_index() \
.sort_values('count_reg', ascending=False)
```

**Выше в таблице находится информация о количестве регистраций на каждый из предметов**

```python
## Чтобы найти процент оттока, надо кол-во ушедших студентов разделить на кол-во всех зарегистрировавшихся
```

```python
## Считаем ли мы оттоком те случаи, когда в одном семестре студент отменил регистрацию,
## а в следующем снова зарегистрировался и прошел предмет? Интуиция подсказывает мне, что да, считаем
```

```python
st_reg['unregistration'] = st_reg.date_unregistration.apply(lambda x: 1 if x != 'NaN' else 0)
```

```python
unreg = st_reg.groupby('code_module').agg({'unregistration': 'sum'}).reset_index()
```

```python
count_reg = pd.merge(count_reg, unreg, on='code_module')
```

```python
count_reg['outflow'] = count_reg['unregistration'] / count_reg['count_reg']
```

```python
count_reg
```

**Таким образом, с самыми большими % оттока являются предметы CCC, DDD и FFF**

### Monetary

```python
monetary = df6.groupby('id_student', as_index=False).agg({'score': 'mean'})
```

```python
monetary['m'] = monetary.score.apply(lambda x: 1 if x < 40 else 3 if x >= 70 else 2)
```

```python
monetary ## Первый шаг сделан
```

### Recency

```python
recency = df6.groupby('id_student', as_index=False).agg({'date_submitted': 'mean'})
```

```python
recency['r'] = recency.date_submitted.apply(lambda x: 1 if x > 245 else 3 if x < 235 else 2)
```

```python
recency
```

### Frequency

```python
## Для подсчета frequency возьмем датафрейм из задания 2
```

```python
df2
```

```python
frequency = df2.groupby('id_student', as_index=False) \
.agg({'result_exam': 'sum'}) \
.rename(columns={'result_exam': 'success'})
```

```python
frequency['f'] = frequency.success.apply(lambda x: 3 if x > 1 else 1 if x < 1 else 2)
```

```python
frequency
```

```python
rfm = pd.merge(recency, frequency, on = 'id_student')
```

```python
rfm = pd.merge(rfm, monetary, on = 'id_student')
```

```python
rfm
```

```python
del rfm['date_submitted']
del rfm['success']
del rfm['score']
```

```python
rfm ## Финишная прямая, радуемся
```

```python
rfm['RFM Score'] = rfm['r'].map(str) + rfm['f'].map(str) + rfm['m'].map(str)
```

```python
rfm
```

## Вот описание сегментов, которые можно выделить:   
**champions** - *Студенты, прошедшие больше одного курса, еще и за каждый набрав по >70 баллов (наши лучшие студенты)*    
**loyal** - *Студенты, прошедшие больше одного курса от 40 до 70 баллов*  
**need attention** - *Студенты, окончившие больше одного курса, набрав от 40 и до 70 баллов*    
**sleep** - *Студенты, прошедшие один курс, пройдя порог*  
**hibernating** - *"Мертвые" студенты, не окончившие ни один курс*  

```python
segt_map = {
    r'[1-3]1': 'hibernating',
    r'1[1-3]': 'hibernating',
    r'22': 'sleep',
    r'32': 'need attention',
    r'23': 'loyal',
    r'33': 'champions',
}

rfm['Segment'] = rfm['m'].map(str) + rfm['f'].map(str)
rfm['Segment'] = rfm['Segment'].replace(segt_map, regex=True)
```

```python
rfm
```

## **Итог:** 
*Мы получили результирующий датафрейм, в котором для каждого студента,  
    который хотя бы раз пытался пройти курс, созданы сегменты.   
    Исходя из того, к какому сегменту относится наш студент, можно придумать  
    подход для повторного привлечения к обучению в нашей школе*  

## **Построим распределение наших студентов по сегментам:**

```python
segments_counts = rfm['Segment'].value_counts().sort_values(ascending=True)

fig, ax = plt.subplots()

bars = ax.barh(range(len(segments_counts)),
              segments_counts,
              color='silver')
ax.set_frame_on(False)
ax.tick_params(left=False,
               bottom=False,
               labelbottom=False)
ax.set_yticks(range(len(segments_counts)))
ax.set_yticklabels(segments_counts.index)

for i, bar in enumerate(bars):
        value = bar.get_width()
        if segments_counts.index[i] in ['champions', 'loyal customers']:
            bar.set_color('silver')
        ax.text(value,
                bar.get_y() + bar.get_height()/2,
                '{:,} ({:}%)'.format(int(value),
                                   int(value*100/segments_counts.sum())),
                va='center',
                ha='left'
               )

plt.show()
```
